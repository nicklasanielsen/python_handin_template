```python
import bs4
import requests

#Go to this website: https://www.komplett.dk/category/21661/gaming/gaming-pc/baerbar
r = requests.get('https://www.komplett.dk/category/21661/gaming/gaming-pc/baerbar')
r.raise_for_status()
#Use BeautifulSoup, to select the name and price for the laptops (hint: look for class="product-price" to begin with).

soup = bs4.BeautifulSoup(r.text, 'html.parser')
prices = soup.select('span[class="product-price-now"]')
names = soup.select('div[class="text-content"] > h2')

laptops = {}

for idx, val in enumerate(names):
    laptops[names[idx].getText()] = prices[idx].getText()

print(laptops)

#Create a mysql table called laptops, with columns id, name, price.

#Insert the laptop data into the laptops table, using a mysql connector and cursor.

# Connect with the MySQL Server
import mysql.connector as mysql

## connecting to the database using 'connect()' method
cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")

cursor = cnx.cursor(prepared=True)

# Insert a laptop
insert_query = """INSERT INTO laptops (name, price) VALUES (%s, %s)"""

for key, value in laptops.items():
    laptop = {key[:45], value}
    val = (key, value,)
    cursor.execute(insert_query, val) 
    cnx.commit()
    
cursor.close()
cnx.close()
print("laptops inserted to db/laptops")
```
